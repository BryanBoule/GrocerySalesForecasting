```python


            

def neural_net_model(X_data,input_dim,nb_hidden, nb_neurons):
    
    tf.set_random_seed(1234)
    
    # layer 1 multiplying and adding bias then activation function
    W_1 = tf.Variable(tf.random_normal([input_dim,nb_neurons]))
    b_1 = tf.Variable(tf.ones([nb_neurons]))[np.newaxis,:]
    tmp = tf.cast(X_data, tf.float32)
    layer_1 = tf.add(tf.matmul(tmp,W_1), b_1)
    layer_1 = tf.nn.relu(layer_1)
    epsilon = 1e-3
    batch_mean1, batch_var1 = tf.nn.moments(layer_1,[0])
    scale1 = tf.Variable(tf.ones([nb_neurons],dtype=tf.float32),dtype=tf.float32)
    beta1 = tf.Variable(tf.zeros([nb_neurons],dtype=tf.float32),dtype=tf.float32)
    layer_1 = tf.nn.batch_normalization(layer_1,batch_mean1,batch_var1,beta1,scale1,epsilon)
    
    W = list(range(1,nb_hidden+2))
    W[0] = W_1
    
    b = list(range(1,nb_hidden+2))
    b[0] = b_1
    
    layer = list(range(1,nb_hidden+2))
    layer[0] = layer_1
    
    batch_mean = list(range(1,nb_hidden+2))
    batch_mean[0] = batch_mean1
    
    batch_var = list(range(nb_hidden+2))
    batch_var[0] = batch_var1
    
    beta = list(range(nb_hidden+2))
    beta[0] = beta1
    
    scale = list(range(nb_hidden+2))
    scale[0] = scale1
    

    
    for i in range(1,nb_hidden+1):
        # layer i multiplying and adding bias then activation function
        W[i] = tf.Variable(tf.random_normal([nb_neurons,nb_neurons]))
        b[i] = tf.Variable(tf.ones([nb_neurons]))[np.newaxis,:]
        
        layer[i]=tf.nn.relu(tf.add(tf.matmul(layer[i-1],W[i]), b[i]))
        
        batch_mean[i], batch_var[i] = tf.nn.moments(layer[i],[0])
        scale[i] = tf.Variable(tf.ones([nb_neurons],dtype=tf.float32),dtype=tf.float32)
        beta[i] = tf.Variable(tf.zeros([nb_neurons],dtype=tf.float32),dtype=tf.float32)
        layer[i] = tf.nn.batch_normalization(layer[i],batch_mean[i],batch_var[i],beta[i],scale[i],epsilon)
        print("HIDDEN LAYER",i)
    
    #Output layer
    W_out = tf.Variable(tf.random_normal([nb_neurons,1]))
    b_out = tf.Variable(tf.ones([1]))[np.newaxis,:]
    output = tf.add(tf.matmul(layer[nb_hidden],W_out), b_out)
    print(" OUTPUT LAYER")
    return output


tf.reset_default_graph()



# On applique ici la stratégie choisie

# Pour la stratégie d'échantillonnage stratifié
if strategie:
    
    #Import des données
    print("DEBUT: IMPORT DES DONNEES")

    #Train
    df_input = pd.read_csv(filepath_or_buffer ="sets/data/little.csv", 
                       engine='c', 
                       infer_datetime_format=True, # to speed-up datetime parsing
                       parse_dates=[1],   
                      )

    #Test
    df_input_test = pd.read_csv(filepath_or_buffer ='sets/data/merged_train_oil_family_cleaned.csv', 
                       engine='c', 
                       infer_datetime_format=True, # to speed-up datetime parsing
                       parse_dates=[1],  
                      )

    print("FIN: IMPORT DES DONNEES")
    
    print("DEBUT: PRETRAITEMENT DES DONNEES")
    df_input_test = df_input_test[df_input_test['date'] > '2017-07-31']

    #Prétraitement des données

    #Empilement des train et test set pour subir le même prétraitement

    df = pd.concat([df_input,df_input_test])

    #Prétraitement des variables

    #On retire la colonne "id" identifiant la ligne uniquement
    df = df.drop(['id'],axis=1)



    #On transforme la date en variables binaires avant de retirer les variables catégorielles 
    #à partir desquelles elles ont été créées
    df['month'] = df['date'].apply(lambda x: x.month)
    df['weekday'] = df['date'].apply(lambda x: x.weekday())
    df['year'] = df['date'].apply(lambda x: x.year) 
    dummy_variables = ['item_nbr', 'store_nbr',
                'month', 'weekday','year','family']

    for var in dummy_variables:
        dummy = pd.get_dummies(df[var], prefix = var, drop_first = False, sparse = True)
        df= df.drop([var], axis = 1)
        df= pd.concat([df, dummy], axis = 1)


    #On sépare le set d'apprentissage et le set de test
    df_train = df.iloc[:len(df_input),:]
    df_test = df.iloc[len(df_input):len(df_input)+len(df_input_test),:]

    
    
# Pour la stratégie de sélection des données par famille
else:
    
    df_input = pd.read_csv(filepath_or_buffer ='sets/data/merged_train_oil_family_cleaned.csv', 
                       engine='c', 
                       infer_datetime_format=True, # to speed-up datetime parsing
                       parse_dates=[1]   
                      )
    # n est la nombre de ligne concernant la famille GROCERY I
    n = len(df_input[(df_input['family']=='GROCERY I')&(df_input['date']<='2017-07-31')])
    
    # On ne garde que la moitié de cette famille car elle est sur-représentée dans les données
    df_to_drop = df_input[(df_input['family']=='GROCERY I')&(df_input['date']<='2017-07-31')].sample(n = int(n/2))
    df = df_input.drop(df_to_drop.index, axis=0)
    
    #On retire la colonne "id" identifiant la ligne uniquement
    df = df.drop(['id'],axis=1)
    
    #On transforme la date en variables binaires avant de retirer les variables catégorielles 
    #à partir desquelles elles ont été créées
    df['month'] = df['date'].apply(lambda x: x.month)
    df['weekday'] = df['date'].apply(lambda x: x.weekday())
    df['year'] = df['date'].apply(lambda x: x.year) 
    dummy_variables = ['item_nbr', 'store_nbr',
                'month', 'weekday','year','family']

    for var in dummy_variables:
        dummy = pd.get_dummies(df[var], prefix = var, drop_first = False, sparse = True)
        df= df.drop([var], axis = 1)
        df= pd.concat([df, dummy], axis = 1)
    
    #On sépare le set d'apprentissage et le set de test
    df_test = df[df['date'] > '2017-07-31']
    df_train = df[df['date'] <= '2017-07-31']

### La stratégie choisie a bien été appliquée    

# On isole les labels de chacun des sets    
y_train = df_train['unit_sales']
df_train = df_train.drop(['unit_sales','date'],axis=1)

y_test = df_test['unit_sales']
df_test = df_test.drop(['unit_sales','date'],axis=1)

print("FIN: PRETRAITEMENT DES DONNEES") 

# LANCEMENT DE LA SESSION DE TENSORFLOW

xs = tf.placeholder(tf.int32)
ys = tf.placeholder(tf.float32)

# Définition de la prédiction
output = neural_net_model(xs,df_train.shape[1],nb_hidden_layers,nb_neurons)

# Définition de la fonction de coup: RMSE
cost = tf.sqrt(tf.reduce_mean(tf.square(output-ys)))

# Optimisation par descente de gradient
train = tf.train.AdamOptimizer(0.001).minimize(cost)

# Définition des batchs
batch_size = 10000
total_len = df_train.shape[0]
total_len_test = df_test.shape[0]
total_batch = int(total_len/batch_size) + 1
total_batch_test = int(total_len_test/batch_size) + 1

# Initialisation de l'erreur test
avg_cost_test = 0

# Initialisation de la liste des prédictions pour le set de test
pred_list = np.array([0]).reshape(1,1)

# Lancement de la session tensorflow    
with tf.Session() as sess:
    print("DEBUT SESSION TENSORFLOW")
    sess.run(tf.global_variables_initializer())
    
    # Boucle sur le nombre d'epoques
    for i in range(nb_epochs):
        
        # Initialisation du coût moyen pour le set d'apprentissage
        avg_cost = 0.
        
        # Boucle sur les batchs
        for j in range(total_batch):
            batch_x = df_train[j*batch_size:(j+1)*batch_size]
            batch_y = y_train[j*batch_size:(j+1)*batch_size][:,np.newaxis]
            
            # Optimisation de la fonction coût par descente de gradient.
            _, c = sess.run([train, cost], feed_dict={xs: batch_x, ys: batch_y})
                        
            # Mise à jour du coût
            avg_cost += c / total_batch
            
        print("époque:",i+1, " erreur d'entraînement: ", avg_cost)
        
    # Initialisation du coût moyen pour le set de test
    avg_cost_test = 0.
        
    # Boucle sur les batchs
    for k in range(total_batch_test):
        batch_test_x = df_test[k*batch_size:(k+1)*batch_size]
        batch_y = y_test[k*batch_size:(k+1)*batch_size][:,np.newaxis]

        # Calcul de la prédiction et de son erreur
        c,pred = sess.run([cost,output], feed_dict={xs: batch_test_x, ys: batch_y})

        # Mise à jour de l'erreur de test
        avg_cost_test += c / total_batch_test

        pred_list = np.concatenate([pred_list, pred], axis=0)
        
    # Liste contenant les prédictions du test set
    result = np.delete(pred_list,0, axis=0)
        
    # Erreur moyenne de test
    print("Erreur moyenne de test: ",avg_cost_test)
```
