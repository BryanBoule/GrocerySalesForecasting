## INTRODUCTION
### Présentation et enjeux

<div style="text-align:justify">Les épiceries doivent faire face à un dilemme de gestion de leurs produits de manière intelligente. Cette gestion demande une bonne connaissance de la clientèle et une bonne prévision d’achats et de ventes. Une prévision de ventes trop optimiste va engendrer des coûts de stockage importants, les épiciers vont se retrouver surchargés avec des produits parfois périssables, source de gaspillage et de perte de revenus. Une prévision de ventes trop pessimiste peut vider les stocks de produits populaires très rapidement, agacer les clients et générer un manque à gagner pour les commerçants. C’est tout l’intérêt de la problématique. L’objectif de ce projet est de créer un modèle capable de prédire le nombre de ventes d’un produit donné. Ainsi, avec les bonnes prévisions, l’épicier pourra commander auprès de ses fournisseurs la bonne quantité dont il a besoin pour optimiser ses revenus et limiter le gaspillage.

Les données de base à notre disposition :

__data.csv__ : contient les quantités de produits vendues par jour / par magasin / par année

__stores.csv__ : contient les emplacements des magasins de l'enseigne implantés en Equateur

__items.csv__ : contient les familles de produits

__oil.csv__ : contient le cours mensuel de l'huile de 2013 au 31 juillet 2018

<br />
<div style="text-align:justify">Les données du problème sont conséquentes, elles pèsent 5Go pour environ 125 millions d'observations et 6 variables explicatives. Afin de gagner en temps de calcul, il est nécessaire de sélectionner une partie des données. Afin de rendre faisable l'apprentissage du réseau sur un temps d'entrainement raisonnable, nous essayerons de nous limiter à des données qui seront les plus représentatives.

Comment sélectionner un échantillon d'observations parmi 125 millions et garder une bonne représentativité de ces données ?

## 2. Traitement des données

On réalise une sélection de produits par famille. La sélection du nombre de produits pris par famille garde la proportionalité des familles pour l'ensemble du dataset.

### 2.1 Sélection selon la famille de produits

```python
# Nombre d'observations par produit
df_nb_obs_per_item = df.item_nbr.value_counts().to_frame().reset_index()
df_nb_obs_per_item = df_nb_obs_per_item.rename(index=str, columns={"index": "item_nbr", "item_nbr": "obs_count"})

# Ajout d'une colonne 'famille' pour chaque produit
df_nb_obs_per_item = pd.merge(df_nb_obs_per_item, df_items, on='item_nbr') 

# Nombre de produits par famille
df_nb_items_per_family = df_nb_obs_per_item.groupby(['family']).size()

# Pourcentage de produits par famille
df_percentage_items_per_family = round(df_nb_items_per_family/sum(df_nb_items_per_family)*100) 
df_percentage_items_per_family = df_percentage_items_per_family.to_frame().reset_index().rename(index=str, columns={0: "item_perc"})

# On trie par ordre décroissant le pourcentage du nombre de produits par famille
df_percentage_items_per_family = df_percentage_items_per_family.sort_values('item_perc', ascending=False)

# On sélectionne les 20 premières familles
families_to_consider = df_percentage_items_per_family[:20]

# On sélectionne le nom des familles
families = families_to_consider.family

# Pourcentage des nombres de produits à prendre par famille
number_items_to_take_per_family = list(map(int, families_to_consider.sort_values('family').item_perc))

# On considère les observations des familles qui nous intéressent
filtred_df_nb_obs_per_item = df_nb_obs_per_item[np.isin(df_nb_obs_per_item['family'], families.tolist())] 

# Regroupement par famille
filtred_df_nb_obs_per_item_group_by_family = filtred_df_nb_obs_per_item.groupby(['family']) 

# Liste des TOP item_nbr pour chaque famille 
list_items = []

# Nombre d'observations du dataframe final
nb_obs = 0 
sorted_families = families_to_consider.sort_values('family').family.tolist()

for i in range(len(families_to_consider)): 
    group_indexes = filtred_df_nb_obs_per_item_group_by_family.groups[sorted_families[i]] #sorted_families[i] : nom de la famille
    for j in group_indexes[:number_items_to_take_per_family[i]]:
        list_items.append(filtred_df_nb_obs_per_item.iloc[j]['item_nbr'])
        nb_obs += filtred_df_nb_obs_per_item.iloc[j]['obs_count']
        
print(nb_obs)    
```

<div style="text-align:justify">Une fois qu'on a récuperé les IDs des produits qui nous intéressent, on récupère toutes les observations correspondantes des données de base.

```python
df_selected = df[np.isin(df['item_nbr'], list_items)]
```

### 4.2 Avec la même proportion d'articles en promotion moyennés sur les 3 dernières années

<div style="text-align:justify">En respectant la proportion du nombre d'observations en promotion sur le nombre d'observations totales et en prenant la valeur moyenne pour chaque année, on peut remplacer les valeurs manquantes équitablement avec la bonne proportion. Seules 2015, 2016 et 2017 seront utilisées pour calculer cette proportion car on dispose également de données manquantes pour l'année 2014.

```python
# Copie des données initiales
df_strategie_2 = df_selected.copy()
```

Calculons la proportion précédemment évoquée.

```python
# On calcule le nombre d'observations qui ne sont pas en promotion par année
nb_obs_pas_en_promotion_par_annee = df_strategie_2[(df_strategie_2['onpromotion']=='False')&(df_strategie_2['year']>2014)].groupby(['year']).count().iloc[:,1]

# On calcule le nombre d'observations qui sont en promotion par année
nb_obs_en_promotion_par_annee = df_strategie_2[(df_strategie_2['onpromotion']=='True')&(df_strategie_2['year']>2014)].groupby(['year']).count().iloc[:,1]

# Calcul de la proportion par année des article qui ne sont pas en promotion
moyenne_nb_obs_pas_en_promotion_par_annee = nb_obs_pas_en_promotion_par_annee/(nb_obs_en_promotion_par_annee + nb_obs_pas_en_promotion_par_annee)
moyenne_nb_obs_pas_en_promotion_par_annee
```

```python
# On calcule la moyenne sur ces trois années
rapport = sum(moyenne_nb_obs_pas_en_promotion_par_annee.tolist())/len(moyenne_nb_obs_pas_en_promotion_par_annee.tolist())
rapport
# 90% d'observations qui ne sont pas en promotion, 10% d'observations qui sont en promotion
```

#### 4.2.1 Année 2013

```python
# On sélectionne toutes les observations de 2013 qui contiennent des valeurs manquantes
tot_2013 = df_strategie_2.loc[(df_strategie_2['year']==2013),'id'].count()

# Nombre d'observations qui seront mises en promotion
nb_true_2013 = int((1-rapport)*tot_2013)
nb_false_2013 = int(rapport*tot_2013)
```

On sélectionne aléatoirement les indices des articles qui seront en promotion en 2013 en respectant la proportion.

```python
# Selection des indexes des observations de la colonne 'onpromotion' qui manquent
index_nan_2013 = df_strategie_2[df_strategie_2['onpromotion']=='NaN'].sample(frac=(1-rapport)).id.tolist()

# On dit que les articles pris en ces indexs seront en promotion
df_strategie_2.loc[np.isin(df_strategie_2['id'], index_nan_2013), ['onpromotion']] = 'True'

# Les autres ne le seront pas
df_strategie_2.loc[(df_strategie_2['onpromotion']=="NaN") & (df_strategie_2['year']==2013),'onpromotion']='False'
```

```python
# Vérification
df_strategie_2.loc[(df_strategie_2['onpromotion']=="NaN") & (df_strategie_2['year']==2013),'id'].count()
```

Il n'y a plus de valeur manquante.

#### 4.2.2 Année 2014

Reprenons ce graphe indiquant le nombre d'observations par année de la colonne 'onpromotion'.

```python
sns.countplot(x = "onpromotion", hue = "year", data = df_selected)
```

```python
# Nombre total d'observations qui sont manquantes 
tot_nan = df_strategie_2.loc[(df_strategie_2['onpromotion']=="NaN") & (df_strategie_2['year']==2014),'id'].count()

# Nombre d'observations qui ne sont pas en promotion avant équilibrage
nb_false_2014_avant_equilibrage = df_strategie_2.loc[(df_strategie_2['onpromotion']=="False") & (df_strategie_2['year']==2014),'id'].count()

# Nombre d'observations qui sont en promotion avant équilibrage
nb_true_2014_avant_equilibrage = df_strategie_2.loc[(df_strategie_2['onpromotion']=="True") & (df_strategie_2['year']==2014),'id'].count()
```

<img src="images/Equations.PNG">

```python
#On résout le système d'équations à 2 inconnues
tot = nb_false_2014_avant_equilibrage+nb_true_2014_avant_equilibrage+tot_nan
nb_true_in_nan_2014 = -(tot-tot_nan) + rapport*nb_false_2014_avant_equilibrage+rapport*tot_nan+(1-rapport)*nb_true_2014_avant_equilibrage
nb_false_in_nan_2014 = tot_nan - nb_true_in_nan_2014
frac_true_2014 = nb_true_in_nan_2014 / (nb_true_in_nan_2014 + nb_false_in_nan_2014)
```

On sélectionne aléatoirement les indices des articles qui seront en promotion en 2014 en respectant la proportion.

```python
index_nan_2014 = df_strategie_2[df_strategie_2['onpromotion']=='NaN'].sample(frac=frac_true_2014).id.tolist()
df_strategie_2.loc[np.isin(df_strategie_2['id'], index_nan_2014), ['onpromotion']] = 'True'
df_strategie_2.loc[(df_strategie_2['onpromotion']=="NaN") & (df_strategie_2['year']==2014),'onpromotion']='False'
```

```python
sns.countplot(x = "onpromotion", hue = "year", data = df_strategie_2)
```

<div style="text-align:justify">La méthode est discutable car on ne prend pas en compte un grand nombre d'années. Cela ne reflète pas forcément l'explication d'une saisonalité à l'échelle annuelle, laquelle pourrait renforcer notre hypothèse en diminuant la valeur de l'écart type ou l'infirmer en l'augmentant. Aussi, les résultats liés à l'année 2016 peuvent fausser ce rapport suite au séisme qui a eu lieu en Avril. Les aides humanitaires et organismes d'aide à la population ont, par leurs dons, contribué à aider les populations, ce qui a permis aux grandes surfaces d'adapter leurs offres promotionnelles.

Il n'y a plus de valeur manquante.

## 5. Gestion des outliers

```python
df_selected = pd.read_csv(filepath_or_buffer ='data/merged_train_oil_family.csv', 
                   engine='c',
                   infer_datetime_format=True, # to speed-up datetime parsing
                   parse_dates=[1],
                   dtype={'onpromotion': object}
                   )
```

```python
df_selected.groupby(['family'])['unit_sales'].sum().nlargest()
```

Il s'avère que sur la période 2013-2018, ce sont les produits de l'**épicerie** qui sont le plus vendus, suivis des **boissons** et des **produits de nettoyage**. Cependant, cela ne nous en dit pas plus sur la distribution des unités vendues car il se peut que des produits soient achetés en moyenne en grande quantité, comparé à d'autres produits.

De ce fait, nous avons choisi de tracer un boxplot de notre étiquette *unit_sales* afin d'observer sa distribution au sein des différentes familles de produits. Nous constatons que des produits sont achetés le plus souvent en grande quantité contrairement à d'autres qui sont souvent ramenés en magasin (ex : viande *vs* produit ménager).

```python
%matplotlib inline
sns.set(style="ticks")

f, ax = plt.subplots(figsize=(25, 8))
ax.set_xscale("log")

sns.boxplot(x="unit_sales", y="family", data=df_selected,
            whis="range", palette="vlag")

ax.xaxis.grid(True)
ax.set(ylabel="")
sns.despine(trim=True, left=True)
```

```python
for i in list(df_selected.family.unique()):
    print(i, ' -> Min : ',
          np.amin(df_selected[df_selected.family == str(i)].unit_sales), ", Q1 : ",
          np.percentile(df_selected[df_selected.family == str(i)].unit_sales, 25), ", Median : ",
          np.percentile(df_selected[df_selected.family == str(i)].unit_sales, 50), ", Q3 : ",
          np.percentile(df_selected[df_selected.family == str(i)].unit_sales, 75), ", P99 : ",
          np.percentile(df_selected[df_selected.family == str(i)].unit_sales, 99), ", Max : ",
          np.amax(df_selected[df_selected.family == str(i)].unit_sales))
```

Comme il était difficile d'extraire certaines informations à partir du boxplot, nous avons décidé de discrétiser la variable *unit_sales*. Nous estimons que cela nous permettra de voir par exemple quels sont les produits qui connaissent le plus souvent des retours en magasins ou encore.
Pour la discrétisation, nous avons choisi de créer 5 modalités:
 - *Retour en magasin* : lorsque le produit est rendu au magasin (*unit_sales* < 0)
 - *[0;Q1[*
 - *[Q1;Q3]*
 - *]Q3;C99[*
 - *[C99;+infini[*

Nous avons créé cette dernière modalité, car en regardant les statistiques descriptives, nous nous sommes aperçus que pour chaque famille de produit, il y avait un écart très important entre le 99ème centile et la valeur maximale.

L’intervalle interquartile est une mesure de la dispersion statistique, étant égale à la différence entre les 3ème quartile et 1er quartile : *IQR = Q3 - Q1*. Elle est similaire à l'écart type ou à la variance mais est beaucoup plus robuste contre les valeurs aberrantes.

```python
Q1 = df_clean.unit_sales.quantile(0.25)
Q3 = df_clean.unit_sales.quantile(0.75)
IQR = Q3 - Q1
```

Pour la détection des *outliers*, nous utilisons donc le score *IQR* afin de créer une base de données ne comportant plus de valeurs aberrantes pour l'apprentissage de notre réseau de neurones.

```python
df_clean = df_clean[((df_clean.unit_sales >= (Q1 - 1.5 * IQR)) & (df_clean.unit_sales <= (Q3 + 1.5 * IQR)))]
```

## 6. Une autre manière de construire les données d'entrainement

### 6.1 Description

<div style="text-align:justify"> Une des stratégies envisagées pour réduire la taille du jeu d'entraînement a été le sous-échantillonnage. Une première façon de procéder est de sous-échantillonner le jeu d'entraînement de manière naïve, en ne gardant qu'un faible pourcentage (dans notre cas 5%) des lignes de la base, lignes sélectionnées de manière aléatoire uniforme. L'échantillonnage uniforme, s'il est assez conséquent, devrait nous permettre de réduire la taille du jeu d'entraînement très simplement tout en conservant les caractéristiques globales de la distribution.
</div>
<br />    

<div style="text-align:justify">Une deuxième manière de procéder est d'effectuer un "stratified sampling", c'est-à-dire de sélectionner 5% de la base mais en conservant la distribution statistique des données. Dans notre cas, il s'agit de garder les mêmes proportions de chaque produit dans le jeu d'entraînement. Cette méthode nous permet d'augmenter les chances que le jeu d'apprentissage ainsi créé conserve la distribution de départ. Il est important de noter qu'habituellement, la stratification s'utilise dans un problème de classification, la variable stratifiée étant la variable à prédire. Ici, nous ne stratifions pas suivant la variable à prédire mais suivant la variable identifiant le produit.
</div>
<br />

<div style="text-align:justify">Nous utilisons pour cela la fonction StratifiedShuffleSplit de Scikit-learn.

## 7. Réseaux de neurones

### 7.1 Implémentation

L'étude peut commencer à partir de ce point, en important simplement le dernier dataframe obtenu.

```python
# CHOIX DE LA STRATEGIE

strategie = 0   # 0 pour la stratégie de sélection des données par famille
                # 1 pour la stratégie d'échantillonnage stratifié
    
# Choix de la structure du réseau de neurones
nb_hidden_layers = 3
nb_neurons = 50

# Choix du nombre d'epochs
nb_epochs = 5
```

neural_net_model est une fonction créant un "feed forward neural net" de nb_hidden couches cachées
qui contiennent chacune nb_neurons (commun à toutes les couches cachées).


Les poids et les biais sont respectivement contenus dans les listes W et b.
Ce sont les variables qui vont être mises à jour pendant l'entraînement.

```python
def neural_net_model(X_data,input_dim,nb_hidden, nb_neurons):
    
    tf.set_random_seed(1234)
    
    # layer 1 multiplying and adding bias then activation function
    W_1 = tf.Variable(tf.random_normal([input_dim,nb_neurons]))
    b_1 = tf.Variable(tf.ones([nb_neurons]))[np.newaxis,:]
    tmp = tf.cast(X_data, tf.float32)
    layer_1 = tf.add(tf.matmul(tmp,W_1), b_1)
    layer_1 = tf.nn.relu(layer_1)
    
    epsilon = 1e-3
    batch_mean1, batch_var1 = tf.nn.moments(layer_1,[0])
    scale1 = tf.Variable(tf.ones([nb_neurons],dtype=tf.float32),dtype=tf.float32)
    beta1 = tf.Variable(tf.zeros([nb_neurons],dtype=tf.float32),dtype=tf.float32)
    layer_1 = tf.nn.batch_normalization(layer_1,batch_mean1,batch_var1,beta1,scale1,epsilon)
    
    W = list(range(1,nb_hidden+2))
    W[0] = W_1
    
    b = list(range(1,nb_hidden+2))
    b[0] = b_1
    
    layer = list(range(1,nb_hidden+2))
    layer[0] = layer_1
    
    batch_mean = list(range(1,nb_hidden+2))
    batch_mean[0] = batch_mean1
    
    batch_var = list(range(nb_hidden+2))
    batch_var[0] = batch_var1
    
    beta = list(range(nb_hidden+2))
    beta[0] = beta1
    
    scale = list(range(nb_hidden+2))
    scale[0] = scale1
    
    for i in range(1,nb_hidden+1):
        # layer i multiplying and adding bias then activation function
        W[i] = tf.Variable(tf.random_normal([nb_neurons,nb_neurons]))
        b[i] = tf.Variable(tf.ones([nb_neurons]))[np.newaxis,:]
        
        layer[i]=tf.nn.relu(tf.add(tf.matmul(layer[i-1],W[i]), b[i]))
        
        batch_mean[i], batch_var[i] = tf.nn.moments(layer[i],[0])
        scale[i] = tf.Variable(tf.ones([nb_neurons],dtype=tf.float32),dtype=tf.float32)
        beta[i] = tf.Variable(tf.zeros([nb_neurons],dtype=tf.float32),dtype=tf.float32)
        layer[i] = tf.nn.batch_normalization(layer[i],batch_mean[i],batch_var[i],beta[i],scale[i],epsilon)
        print("HIDDEN LAYER",i)
    
    #Output layer
    W_out = tf.Variable(tf.random_normal([nb_neurons,1]))
    b_out = tf.Variable(tf.ones([1]))[np.newaxis,:]
    output = tf.add(tf.matmul(layer[nb_hidden],W_out), b_out)
    print(" OUTPUT LAYER")
    return output
```

Dans ce notebook, nous avons uniquement lancé le réseau de neurones pour la stratégie de sélection des données par famille, faute de temps et aussi parce que ça nous a permis d'avoir de meilleurs résultats que ceux obtenus par la stratégie d'échantillonnage stratifié. Par la suite, nous analyserons les résultats obtenus par notre réseau de neurones.

```python
tf.reset_default_graph()

# On applique ici la stratégie choisie
# Pour la stratégie d'échantillonnage stratifié
if strategie:
    
    #Import des données
    print("DEBUT: IMPORT DES DONNEES")

    #Train
    df_input = pd.read_csv(filepath_or_buffer ="data/strat_trained_merged.csv", 
                       engine='c', 
                       infer_datetime_format=True, # to speed-up datetime parsing
                       parse_dates=[1],   
                      )

    #Test
    df_input_test = pd.read_csv(filepath_or_buffer ='data/merged_train_oil_family_cleaned.csv', 
                       engine='c', 
                       infer_datetime_format=True, # to speed-up datetime parsing
                       parse_dates=[1],  
                      )

    print("FIN: IMPORT DES DONNEES")
    
    print("DEBUT: PRETRAITEMENT DES DONNEES")
    df_input_test = df_input_test[df_input_test['date'] > '2017-07-31']

    #Prétraitement des données

    #Empilement des train et test set pour subir le même prétraitement

    df = pd.concat([df_input,df_input_test])

    #Prétraitement des variables

    #On retire la colonne "id" identifiant la ligne uniquement
    df = df.drop(['id'],axis=1)



    #On transforme la date en variables binaires avant de retirer les variables catégorielles 
    #à partir desquelles elles ont été créées
    df['month'] = df['date'].apply(lambda x: x.month)
    df['weekday'] = df['date'].apply(lambda x: x.weekday())
    df['year'] = df['date'].apply(lambda x: x.year) 
    dummy_variables = ['item_nbr', 'store_nbr',
                'month', 'weekday','year','family']

    for var in dummy_variables:
        dummy = pd.get_dummies(df[var], prefix = var, drop_first = False, sparse = True)
        df= df.drop([var], axis = 1)
        df= pd.concat([df, dummy], axis = 1)


    #On sépare le set d'apprentissage et le set de test
    df_train = df.iloc[:len(df_input),:]
    df_test = df.iloc[len(df_input):len(df_input)+len(df_input_test),:]

    
    
# Pour la stratégie de sélection des données par famille
else:
    # n est la nombre de ligne concernant la famille GROCERY I
    n = len(df_clean[(df_clean['family']=='GROCERY I')&(df_clean['date']<='2017-07-31')])
    
    # On ne garde que la moitié de cette famille car elle est sur-représentée dans les données
    df_to_drop = df_clean[(df_clean['family']=='GROCERY I')&(df_clean['date']<='2017-07-31')].sample(n = int(n/2))
    df_input = df_clean.drop(df_to_drop.index, axis=0)
    
    #On retire la colonne "id" identifiant la ligne uniquement
    df_input = df_input.drop(['id'],axis=1)
    
    #On transforme la date en variables binaires avant de retirer les variables catégorielles 
    #à partir desquelles elles ont été créées
    df_input['month'] = df_input['date'].apply(lambda x: x.month)
    df_input['weekday'] = df_input['date'].apply(lambda x: x.weekday())
    df_input['year'] = df_input['date'].apply(lambda x: x.year) 
    dummy_variables = ['onpromotion', 'item_nbr', 'store_nbr',
                'month', 'weekday','year','family']

    for var in dummy_variables:
        dummy = pd.get_dummies(df_input[var], prefix = var, drop_first = False, sparse = True)
        df_input= df_input.drop([var], axis = 1)
        df_input= pd.concat([df_input, dummy], axis = 1)
    
    #On sépare le set d'apprentissage et le set de test
    df_test = df_input[df_input['date'] > '2017-07-31']
    df_train = df_input[df_input['date'] <= '2017-07-31']

### La stratégie choisie a bien été appliquée    

# On isole les labels de chacun des sets    
y_train = df_train['unit_sales']
df_train = df_train.drop(['unit_sales','date'],axis=1)

y_test = df_test['unit_sales']
df_test = df_test.drop(['unit_sales','date'],axis=1)

print("FIN: PRETRAITEMENT DES DONNEES") 

# LANCEMENT DE LA SESSION DE TENSORFLOW

xs = tf.placeholder(tf.int32)
ys = tf.placeholder(tf.float32)

# Définition de la prédiction
output = neural_net_model(xs,df_train.shape[1],nb_hidden_layers,nb_neurons)

# Définition de la fonction de coup: RMSE
cost = tf.sqrt(tf.reduce_mean(tf.square(output-ys)))

# Optimisation par descente de gradient
train = tf.train.AdamOptimizer(0.001).minimize(cost)

# Définition des batchs
batch_size = 10000
total_len = df_train.shape[0]
total_len_test = df_test.shape[0]
total_batch = int(total_len/batch_size) + 1
total_batch_test = int(total_len_test/batch_size) + 1

# Initialisation de la liste des prédictions pour le set de test
pred_list = np.array([0]).reshape(1,1)

# Lancement de la session tensorflow    
with tf.Session() as sess:
    print("DEBUT SESSION TENSORFLOW")
    sess.run(tf.global_variables_initializer())
    
    # Boucle sur le nombre d'epoques
    for i in range(nb_epochs):
        
        # Initialisation du coût moyen pour le set d'apprentissage
        avg_cost = 0.
        
        # Boucle sur les batchs
        for j in range(total_batch):
            batch_x = df_train[j*batch_size:(j+1)*batch_size]
            batch_y = y_train[j*batch_size:(j+1)*batch_size][:,np.newaxis]
            
            # Optimisation de la fonction coût par descente de gradient.
            _, c = sess.run([train, cost], feed_dict={xs: batch_x, ys: batch_y})
                        
            # Mise à jour du coût
            avg_cost += c / total_batch
            
        print("époque:",i+1, " erreur d'entraînement: ", avg_cost)
        
    # Initialisation du coût moyen pour le set de test
    avg_cost_test = 0.
        
    # Boucle sur les batchs
    for k in range(total_batch_test):
        batch_test_x = df_test[k*batch_size:(k+1)*batch_size]
        batch_y = y_test[k*batch_size:(k+1)*batch_size][:,np.newaxis]

        # Calcul de la prédiction et de son erreur
        c,pred = sess.run([cost,output], feed_dict={xs: batch_test_x, ys: batch_y})

        # Mise à jour de l'erreur de test
        avg_cost_test += c / total_batch_test

        pred_list = np.concatenate([pred_list, pred], axis=0)
        
    # Liste contenant les prédictions du test set
    result = np.delete(pred_list,0, axis=0)
        
    # Erreur moyenne de test
    print("Erreur moyenne de test: ",avg_cost_test)
```

Dans cette figure, nous affichons le min et le max d'erreur pour chaque famille.

```python
categories = np.array(range(5))
array_by_category = [np.sum((error/y_test[:,np.newaxis] >= i*0.2) & (error/y_test[:,np.newaxis] < (i+1)*0.2))/len(y_test)*100 for i in categories]
array_by_category = array_by_category + [np.sum(error/y_test[:,np.newaxis] >= 1)/len(y_test)*100]
```
